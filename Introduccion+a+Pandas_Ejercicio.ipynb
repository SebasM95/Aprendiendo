# Renombrar Columnas

```python
# renombrar columnas "User Rating," "Critic Rating" y "Average Rating" a "UR," "CR" y "AR" Luego actualizar el dataframe con el parametro "inplace"
db_bestseller1=db_bestseller = db_bestseller.rename(columns={'User Rating': 'UR',
                                             'Critic Rating': 'CR',
                                             'Average Rating': 'AR'})
db_bestseller
```

```python
# mostrar 5 primeras filas
db_bestseller1.head(5)
```

```python
# seleccionar solo columnas "Name", "Author", "UR", "CR", "AR" y "Year" y actualizar el dataframe
db_bestseller_orden= db_bestseller1.loc[:, ['Name', 'Author', 'UR', 'CR', 'AR', 'Year']]
db_bestseller_orden
```

# Ordenar el dataframe

```python
# ordenar el dataframe en forma descendente segun las columnas "UR" y "CR"
db_bestseller_orden.sort_values(['UR','CR'],ascending=False,
                           inplace=True)
db_bestseller_orden
```
