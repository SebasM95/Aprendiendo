```python
#Siempre lo primero que debo hacer es importar pandas y numpy
import pandas as pd
import numpy as np
```

```python
#Ahora subismo el archivo desde google drive, ya no tenemos que loguearnos de nuevo
df_estudiantes=pd.read_csv('/content/drive/MyDrive/StudentsPerformance (1) (1).csv')
df_estudiantes
```

```python
#Sigamos con la creación de de columnas usando el método assign o el método insert
#Si quiero agregar varias columnas uso assign y tambien si quiero sobreescribir una columna
```

```python
#Creación de notas de fisica y filosofia aleatorias con un rango de 1 a 100 para 100 filas
filosolofia_score=np.random.randint(1,100,size=1000)
fisica_score=np.random.randint(1,100,size=1000)
```

```python
#Ahora le pondre un indice a cada fila para que sepa en donde va el dataframe
filosofia_nota=pd.Series(filosolofia_score, index=np.arange(0,1000))
fisica_nota=pd.Series(fisica_score, index=np.arange(0,1000))
```

```python
#Ahora lo que voy a hacer es poner las nueva columnas a el dataframe usando assign
df_estudiantes.assign(filosolofia_score=filosofia_nota,fisica_score=fisica_nota)
```

```python
#El metodo instert es diferente por que me permite especificar donde voy a colocar la columna
#Donde yo voy a colocar una columna
#Al usar el metodo insert debemos tener cuidado por que no se devuelve si no que el cambio queda realizado de una vez
df_estudiantes.insert(1, 'ED_SCORE',fisica_nota)
df_estudiantes
```
